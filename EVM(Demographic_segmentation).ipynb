```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# Copy and encode the categorical variables
df_encoded = df.copy()
for col in ['Profession', 'Marrital Status', 'Education', 'Personal loan', 'House Loan', 'Wife Working']:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Compute correlation (excluding 'Make')
corr = df_encoded.drop('Make', axis=1).corr()

# High-resolution, larger plot
plt.figure(figsize=(15, 12), dpi=120)  # bigger and clearer
sns.heatmap(
    corr,
    annot=True,
    cmap='mako',
    fmt=".2f",
    annot_kws={"size": 10},
    linewidths=0.5,
    linecolor='white',
    square=True,
    cbar_kws={"shrink": 0.8}
)
plt.title("Correlation Heatmap", fontsize=18)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()
```

We observe high correlation between total salary (salary + wife salary) and price.

```python
dfe = df.iloc[:,df.columns!='Make']
dfe.head()
```

```python
dfe['Profession'] = le.fit_transform(dfe['Profession'])
dfe['Marrital Status'] = le.fit_transform(dfe['Marrital Status'])
dfe['Education'] = le.fit_transform(dfe['Education'])
dfe['Personal loan'] = le.fit_transform(dfe['Personal loan'])
dfe['House Loan'] = le.fit_transform(dfe['House Loan'])
dfe['Wife Working'] = le.fit_transform(dfe['Wife Working'])
```

```python
dfe.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
dfe_scaled = StandardScaler().fit_transform(dfe)
dfe_scaled = pd.DataFrame(dfe_scaled,columns=['Age', 'Profession', 'Marrital Status', 'Education', 'No of Dependents',
                                          'Personal loan', 'House Loan', 'Wife Working', 'Salary', 'Wife Salary',
                                          'Total Salary','Price'])
x = dfe_scaled.to_numpy()
dfe_scaled
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',
                    max_iter = 300, n_init = 10,random_state = 0)
    kmeans.fit(dfe_scaled)
    wcss.append(kmeans.inertia_)
sns.set_style("darkgrid")
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Square') # Within cluster sum of squares
plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++',
                    max_iter = 300, n_init = 10,random_state = 42)
kmeans.fit(dfe_scaled)
```

```python
y = kmeans.predict(dfe_scaled)
y_df = pd.DataFrame(y,columns=['Class'])
final_data = pd.concat([df,y_df],axis=1)
final_data
```

```python
sns.pairplot(final_data,x_vars = ['Age','Salary', 'Wife Salary','Total Salary','Price'], y_vars = ['Age','Salary', 'Wife Salary','Total Salary','Price'], hue='Class', palette= 'bright')
```

```python
#k = 3
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                    max_iter = 300, n_init = 10,random_state = 42)
kmeans.fit(dfe_scaled)
```

```python
y = kmeans.predict(dfe_scaled)
y_df = pd.DataFrame(y,columns=['Class'])
final_data = pd.concat([df,y_df],axis=1)
```

```python
sns.pairplot(final_data,x_vars = ['Age','Salary', 'Wife Salary','Total Salary','Price'], y_vars = ['Age','Salary', 'Wife Salary','Total Salary','Price'], hue='Class', palette= 'bright')
```

Thus, we see that we can get a more meaningful analysis for k = 3

```python
# Load the original dataset
df = pd.read_csv('/content/4-wheeler-EV-carwale.csv')

# Simulate 'PriceEuro' using 'rating' (since no price column is available)
df['inr(10e3)'] = df['rating'] * 0.08320

# Simulate 'RapidCharge' column as 'Yes'/'No' randomly
np.random.seed(0)
df['RapidCharge'] = np.random.choice(['Yes', 'No'], size=len(df))

# Convert 'RapidCharge' to numeric (Yes  1, No  0)
df['RapidCharge'].replace({'No': 0, 'Yes': 1}, inplace=True)

# Display in your desired format
output_df = df[['model_name', 'rating', 'inr(10e3)', 'RapidCharge']]
print(output_df.head())
```

```python
# Extract brand from model_name
df['Brand'] = df['model_name'].apply(lambda x: str(x).split()[0].capitalize())
df['Efficiency_WhKm'] = df['Fuel Economy'] * 10
plt.figure(figsize=(20, 5))
sns.barplot(x='Brand', y='Efficiency_WhKm', data=df, palette='rainbow')
plt.grid(axis='y')
plt.title('Analysis based on efficiency achieved by a brand:')
plt.xlabel('Brand')
plt.ylabel('Efficiency')
plt.xticks(rotation=45)
plt.show()
```

```python

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Efficiency_WhKm'], df['Value for Money'] * 1000, color='blue', alpha=0.7)
plt.xlabel('Efficiency (Wh/km)')
plt.ylabel('Value for Money (scaled 1000)')
plt.title('Efficiency vs Value for Money of EVs')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Thus, we see that highly efficient models are available at low prices.
